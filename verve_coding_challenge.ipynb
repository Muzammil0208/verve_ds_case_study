--- Part One ---  
Find the two entries that sum to 2020 and then multiply those two numbers together.  

You can run the code by executing the cell.

### Approach
1. seen_numbers: This set is used to store the numbers weâ€™ve already encountered. 
2. For each number in the data :  
   a) Calculate the complement , the number needed to make the sum 2020 when added to the current number.  
   b) Check if complement exists in the set. If it does this means we have already seen the number that can pair with current number.  
   c) If complement is found, store the pair and the product in the results list.  
   d) Add the current number to the set for future complement checks. By trying to identify if there are multiple pairs.  

3. After iterating over the entire dataset, the results are printed, showing the pairs that sum to 2020 and their corresponding products. 

```python
results = []
total_sum = 2020
seen_numbers = set()

for num in data:
    complement = total_sum - num

    if complement in seen_numbers:
        results.append((num, complement, num*complement))

    seen_numbers.add(num)

for result in results:
    print(f"Two Numbers: {result[0]} and {result[1]}")
    print(f"Product of the Numbers: {result[2]}")
```

--- Part Two ---  
Find three numbers in your dataset that meet the same criteria. 

### Approach
1. There are two loops which are required to find the three numbers.
2. In the outer loop,   
   a) Iterate through the each number which picks a number that will be the first number in the potential list. 
   b) Calculate the target sum, which gives the remaining sum that needs to be made by two numbers. 

3. In the inner loop,  
   a) Iterate through all the subsequent elements starting from index i+1 to avoid checking the same pair twice.  
   b) Calculate the complement which makes the sum to 2020.

4. If the complement is found in the set, then the first number, inner loop number and complement make up for the sum 2020.

5. If the complement is not found, the number is added to the seen numbers set.

6. After all the loops are finished the three numbers and their product is printed.


```python
# List to store the results
results_three_numbers = []

#Outer loop
for i in range(len(data)):
    seen_numbers = set()
    target_sum = 2020 - data[i]
    
    #inner loop
    for j in range(i + 1, len(data)):
        complement = target_sum - data[j]
        
        if complement in seen_numbers:
            results_three_numbers.append((data[i], data[j], complement, data[i] * data[j] * complement))
        
        seen_numbers.add(data[j])

for result in results_three_numbers:
    print(f"Three Numbers: {result[0]}, {result[1]}, and {result[2]}")
    print(f"Product of three numbers: {result[3]}")
```
